```python
import pandas as pd
```

```python
#Load the data from provided files: ‘Patient Info.xlsx’ and ‘Partner_Partner_Mapping.xlsx’.
patient_info = pd.read_excel('Patient_info.xlsx')
partner_partner_mapping = pd.read_excel('Partner_Partner_Mapping.xlsx')
```

```python
#Join both data frames based on Patient_ID. 
join_data = pd.merge(patient_info, partner_partner_mapping, on='Patient_ID')
join_data
```

```python
#Get patient information of those patients who do not have a partner.
data = pd.merge(patient_info, partner_partner_mapping, on='Patient_ID', how='left')
no_partner_patients = data[data["Partner_ID"].isna()]
no_partner_patients
```

```python
# Get patient information of only primary patients.
primary_patient_info = data[["Patient_ID", "Patient_Name", "Patient_Gender", "Patient_Phone", "Province", "Created_at"]]
primary_patient_info
```

```python
#Create a dataframe/dictionary (whichever you prefer) to show how many patients are from each unique province
unique_province_patient_count = pd.DataFrame(data.groupby(["Province"]).Patient_ID.count()).reset_index()
unique_province_patient_count = unique_province_patient_count.rename(columns={'Patient_ID': 'Patient_count'})
unique_province_patient_count
```

```python
#Create a table with the number of referrals each month.
referral_per_month = pd.DataFrame(patient_info[["Created_at"]].groupby(pd.Grouper(key='Created_at', freq='1M')).size().reset_index(name='referral_per_month'))
referral_per_month
```
